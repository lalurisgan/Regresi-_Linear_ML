```python
# Mengimpor library numpy dan memberinya alias np.
# Numpy digunakan untuk komputasi numerik, terutama untuk array dan matriks.
import numpy as np

# Mengimpor library pandas dan memberinya alias pd.
# Pandas digunakan untuk manipulasi dan analisis data, terutama dengan DataFrame.
import pandas as pd

# Mengimpor modul pyplot dari library matplotlib dan memberinya alias plt.
# Matplotlib digunakan untuk visualisasi data, dan pyplot menyediakan fungsi untuk membuat plot.
from matplotlib import pyplot as plt

# Mengimpor kelas LinearRegression dari modul sklearn.linear_model.
# Scikit-learn (sklearn) digunakan untuk machine learning, dan LinearRegression adalah model regresi linear.
from sklearn.linear_model import LinearRegression
```

```python
# Membaca data dari file CSV 'R01_students_performance.csv'
# dan menyimpannya dalam variabel bernama 'dataset'.
# pd.read_csv adalah fungsi dari library pandas untuk membaca data dari file CSV.
dataset = pd.read_csv('R01_students_performance.csv')

# Menampilkan isi dari variabel 'dataset'.
# Ini akan mencetak DataFrame yang berisi data dari file CSV.
dataset
```

```python
# Mengubah nama kolom pada DataFrame 'dataset'.
# 'dataset.columns' mengakses atribut kolom dari DataFrame.
# ['x', 'y'] adalah daftar nama kolom baru.
dataset.columns = ['x', 'y']
# Menampilkan DataFrame 'dataset' dengan nama kolom yang sudah diubah.
dataset
```

```python
dataset.corr() # dataset.corr() digunakan untuk menghitung korelasi antar kolom
# di DataFrame 'dataset'.
# Secara default, menggunakan metode
```

```python
# Mengambil data kolom 'x' dan mengubahnya menjadi array NumPy.
x_train = dataset['x'].values
# Menambahkan dimensi baru ke array x_train agar sesuai dengan format input model machine learning.
# np.newaxis menyisipkan axis baru pada posisi yang ditentukan.
x_train = x_train[:, np.newaxis]
# Mengambil data kolom 'y' dan mengubahnya menjadi array NumPy.
# Ini adalah data target yang akan diprediksi oleh model.
y_train = dataset['y'].values
```

```python
# Membuat objek model regresi linear dan menyimpannya dalam variabel 'lm'.
lm = LinearRegression()
# Melatih model regresi linear 'lm' menggunakan data pelatihan.
# x_train: Data fitur (variabel independen) untuk pelatihan.
# y_train: Data target (variabel dependen) yang sesuai dengan x_train.
# Proses ini menyesuaikan parameter model agar dapat memprediksi y_train berdasarkan x_train.
lm.fit(x_train, y_train)
```

```python
# Membuat data uji untuk prediksi, berisi nilai StudyHours
x_test = [[0.5], [1.0], [1.5], [2.0], [2.5], [3.0]]
# Memprediksi ExamScores menggunakan model dan data uji
predictions = lm.predict(x_test)

# Menampilkan koefisien (StudyHours) dari model regresi
print('StudyHours :', lm.coef_)
# Menampilkan intercept (ExamScores) dari model regresi
print('ExamScores :', lm.intercept_)
```

```python
y_pred_train = lm.predict(x_train)
dfc = pd.DataFrame({'x': dataset['x'], 'y': y_pred_train}) # y_pred_train = lm.predict(x_train)
# Variabel y_pred_train menyimpan hasil prediksi nilai ExamScores
# berdasarkan data StudyHours (x_train) menggunakan model lm
# yang sudah dilatih sebelumnya.

# dfc = pd.DataFrame({'x': dataset['x'], 'y': y_pred_train})
# Membuat DataFrame baru bernama dfc yang berisi dua kolom:
#   - 'x': berisi data StudyHours asli dari dataset.
#   - 'y': berisi nilai ExamScores yang diprediksi (y_pred_train).
# DataFrame ini dapat digunakan untuk analisis lebih lanjut,
# misalnya membandingkan nilai prediksi dengan nilai asli
# untuk melihat kinerja model.
```

```python
# Membuat scatter plot dari data asli.
# dataset['x'] : Data StudyHours (sumbu x).
# dataset['y'] : Data ExamScores (sumbu y).
# label="Data asli" : Memberi label "Data asli" pada scatter plot.
plt.scatter(dataset['x'], dataset['y'], label="Data asli")

# Membuat garis regresi.
# dataset['x'] : Data StudyHours (sumbu x).
# dfc['y'] : Data ExamScores hasil prediksi (sumbu y).
# color='Yellow' : Memberi warna kuning pada garis.
# linewidth=1 : Mengatur ketebalan garis.
# label="Regression Line" : Memberi label "Regression Line" pada garis.
plt.plot(dataset['x'], dfc['y'], color='Yellow', linewidth=1, label="Regression Line")

# Memberi label pada sumbu x.
plt.xlabel('StudyHours')

# Memberi label pada sumbu y.
plt.ylabel('ExamScores')

# Menampilkan legenda.
plt.legend()

# Menampilkan plot.
plt.show()
```

```python
# Mendefinisikan variabel y_asli dengan nilai [0.8, 1.7, 0.4].
# Kemungkinan berisi nilai ExamScores asli, tapi tidak digunakan selanjutnya.
y_asli = [0.8, 1.7, 0.4]

# Memprediksi ExamScores menggunakan model lm dengan data StudyHours baru.
# [[0.27], [0.60], [0.23]] adalah data StudyHours untuk prediksi.
# Hasil prediksi disimpan dalam variabel y_pred_hasil.
y_pred_hasil = lm.predict([[0.27], [0.60], [0.23]])

# Menampilkan hasil prediksi ke layar.
print("Predicted values for [0.27], [0.60], [0.23]:", y_pred_hasil)
```

```python
# Mengimpor fungsi r2_score dari modul sklearn.metrics
# Fungsi ini digunakan untuk menghitung nilai R-squared.
from sklearn.metrics import r2_score

# Menghitung nilai R-squared dan menyimpannya dalam variabel 'akurasi'.
# y_asli: Nilai aktual/observasi.
# y_pred_hasil: Nilai prediksi dari model.
akurasi = r2_score(y_asli, y_pred_hasil)

# Menampilkan nilai R-squared dalam bentuk persentase.
# Mengalikan dengan 100 untuk mengubahnya menjadi persentase.
print(akurasi*100)
```
